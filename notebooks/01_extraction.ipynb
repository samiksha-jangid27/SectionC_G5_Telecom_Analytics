```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set aesthetic parameters for plotting
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
```

## 1. Load the Dataset
We will load the `telecom_churn_raw_dataset.csv` file.

```python
file_path = "../data/raw/telecom_churn_raw_dataset.csv"

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print(f"Dataset loaded successfully! Shape: {df.shape}")
else:
    print(f"Error: File not found at {file_path}")
```

## 2. Initial Inspection
Let's look at the first few rows and the column data types.

```python
# Display first 5 rows
display(df.head())

# Display column information and types
df.info()
```

## 3. Data Quality Assessment
We need to check for:
- **Missing Values**: Are there any nulls?
- **Duplicates**: Are there repeated customer IDs?
- **Anomalies**: Negative values in columns like `data_used` or `calls_made`.

```python
print("--- Missing Values ---")
print(df.isnull().sum())

print("\n--- Duplicate Rows ---")
print(f"Total duplicates: {df.duplicated().sum()}")

print("\n--- Summary Statistics (Numerical) ---")
display(df.describe())
```

## 4. Key Observations & Issues Identified
Based on the `describe()` output, check for:
- Minimum values of `calls_made`, `sms_sent`, and `data_used` (should not be negative).
- Unique count of `customer_id` vs total rows.

```python
# Check for negative values in specific columns
numerical_cols = ['calls_made', 'sms_sent', 'data_used', 'estimated_salary', 'age']
for col in numerical_cols:
    neg_count = (df[col] < 0).sum()
    if neg_count > 0:
        print(f"Warning: Column '{col}' contains {neg_count} negative values.")
```
