# 05. Final Export & Tableau Preparation

## Overview
The final stage of our pipeline involves formatting the data for visualization tools like Tableau or PowerBI.

## Objectives:
1. **Semantic Renaming**: Convert technical column names to user-friendly titles.
2. **Final Export**: Generate a clean CSV in the `tableau/` directory.

```python
import pandas as pd
import os

# Load processed data
df = pd.read_csv("../data/processed/telecom_churn_cleaned.csv")

# Column Map for Visualization
tableau_columns = {
    'telecom_partner': 'Telecom Partner',
    'gender': 'Gender',
    'age': 'Customer Age',
    'state': 'Region/State',
    'city': 'City',
    'estimated_salary': 'Annual Income',
    'calls_made': 'Monthly Calls',
    'sms_sent': 'Monthly SMS',
    'data_used': 'Data Usage (GB)',
    'churn': 'Churn Flag',
    'age_group': 'Age Segment',
    'usage_score': 'Total Engagement Score'
}

df_final = df.rename(columns=tableau_columns)

# Create final summary features
df_final['Churn Text'] = df_final['Churn Flag'].map({1: 'Yes', 0: 'No'})

print(f"Finalizing {len(df_final)} records for visualization.")
```

## Exporting to Tableau Directory

```python
output_dir = "../tableau"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

final_path = os.path.join(output_dir, "telecom_tableau_export.csv")
df_final.to_csv(final_path, index=False)

print(f"[SUCCESS] Final dataset exported to: {final_path}")
```
